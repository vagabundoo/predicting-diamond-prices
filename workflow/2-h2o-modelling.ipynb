```python
import h2o
h2o.init()
```

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
diamond_path = '../cleaned-data/01-diamonds-train-cl.csv'
diam_df = h2o.import_file(path=diamond_path).drop(0, axis=1)
diam_df.describe()
```

```python
diam_df_train = diam_df.drop(0, axis=1)
print(list(enumerate(diam_df_train.columns)))
diam_df_train
```

```python
gbm_regressor = H2OGradientBoostingEstimator(distribution="gaussian",
ntrees=10, max_depth=4, min_rows=2, learn_rate=0.2)
```

```python
gbm_regressor.train(x=diam_df_train.columns[:-1], y='price', training_frame=diam_df_train)
```

```python
gbm_regressor
```

```python
train, test = diam_df.split_frame([0.8], seed=1234)
gbm_regressor.train(x=train.columns[:-1], y='price', training_frame=train)
```

```python
gbm_regressor.predict(test).shape
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
diam_X = diam_df.col_names[:-1]     #last column is price, our desired response variable 
diam_y = diam_df.col_names[-1]
```

```python
diam_df
```

```python
diam_forest_v1 = H2ORandomForestEstimator(
    model_id="rf_covType_v1",
    ntrees=100,
    stopping_rounds=2,
    score_each_iteration=True,
    seed=1000000)
diam_forest_v1.train(diam_X, diam_y, training_frame=train, validation_frame=test)
```

```python
diam_forest_v1
```

```python
# We remove the variables that contribute the least to the model

diam_forest_v2 = H2ORandomForestEstimator(
    model_id="rf_covType_v1",
    ntrees=100,
    stopping_rounds=2,
    score_each_iteration=True,
    seed=1000000)
diam_forest_v2.train(['carat', 'y','x','z','clarity','color','depth'], diam_y, training_frame=train, validation_frame=test)
```

```python
diam_forest_v1
```

```python
diam_forest_v3 = H2ORandomForestEstimator(
    model_id="rf_covType_v1",
    ntrees=10,
    stopping_rounds=2,
    score_each_iteration=True,
    seed=1000000)
diam_forest_v3.train(['carat', 'y','x','z','clarity','color'], diam_y, training_frame=train, validation_frame=test)
```

```python
diam_forest_v3.rmse
```
